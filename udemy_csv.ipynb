# Visuals To Better Understand Dataset:-

```python
subject_counts = udemy['subject'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(subject_counts, labels=subject_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Courses by Subject')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
top_courses = udemy.sort_values('num_subscribers', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_courses['course_title'], top_courses['num_subscribers'], color='skyblue')
plt.xticks(rotation=75, ha='right')  # Rotate titles for readability
plt.xlabel('Course Title')
plt.ylabel('Number of Subscribers')
plt.title('Top 10 Courses by Number of Subscribers')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Optional: Display only top 10 courses for readability
top_courses = udemy.head(10)  # or use sort_values() if you want most popular ones

plt.figure(figsize=(12, 6))
sns.barplot(data=top_courses, x='course_title', y='num_subscribers', hue='subject', dodge=False)
plt.xticks(rotation=75, ha='right')
plt.xlabel('Course Title')
plt.ylabel('Number of Subscribers')
plt.title('Courses by Subject and Number of Subscribers')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Histogram for number of subscribers
axs[0].hist(udemy['num_subscribers'], bins=30, color='skyblue')
axs[0].set_title('Distribution of Subscribers')
axs[0].set_xlabel('Subscribers')
axs[0].set_ylabel('Frequency')

# Histogram for number of reviews
axs[1].hist(udemy['num_reviews'], bins=30, color='lightgreen')
axs[1].set_title('Distribution of Reviews')
axs[1].set_xlabel('Reviews')

# Histogram for number of lectures
axs[2].hist(udemy['num_lectures'], bins=30, color='salmon')
axs[2].set_title('Distribution of Lectures')
axs[2].set_xlabel('Lectures')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.hist(udemy['num_subscribers'], bins=30, alpha=0.5, label='Subscribers')
plt.hist(udemy['num_reviews'], bins=30, alpha=0.5, label='Reviews')
plt.hist(udemy['num_lectures'], bins=30, alpha=0.5, label='Lectures')
plt.legend()
plt.title('Distribution of Subscribers, Reviews, and Lectures')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
